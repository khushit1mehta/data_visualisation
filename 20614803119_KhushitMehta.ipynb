#  XGBR vs LINEAR REGRESSION ON BOSTON DATASET

# IMPORTING ALL THE LIBRARIES

```python
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 
import numpy as np
from sklearn import datasets 
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import mean_squared_error
from sklearn import datasets
```

```python
#IMPORTING THE DATASET
boston = datasets.load_boston()
```

```python
print(boston.data.shape)
print(boston.feature_names)
```

```python
y=(boston.target)
x=(boston.data)
x.shape
df=pd.DataFrame(x)
df.columns = boston.feature_names
```

```python
df.head(15)
```

```python
df['Price'] = boston.target
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python

xtrain, xtest, ytrain, ytest = model_selection.train_test_split(x, y, test_size=0.30,random_state=2)
```

```python
print("xtrain shape : ", xtrain.shape)
print("xtest shape  : ", xtest.shape)
print("ytrain shape : ", ytrain.shape)
print("ytest shape  : ", ytest.shape)
```

```python
algoR = LinearRegression()
algoR.fit(xtrain,ytrain)

algoX = xgb.XGBRegressor()
algoX.fit(xtrain,ytrain)
```

```python
y_R_pred = algoR.predict(xtest)
y_X_pred = algoX.predict(xtest)
```

```python
score_R = algoR.score(xtrain, ytrain)   
print("Training score of Linear Regression: ", score_R)
score_X = algoX.score(xtrain, ytrain)   
print("Training score of XGBREGRESSOR: ", score_X)
```

```python
# - cross validataion OF LR
scores_R = cross_val_score(algoR, xtrain, ytrain, cv=5)
print("Mean cross-validation score LR: %.2f" % scores_R.mean())

# - cross validataion OF XGBR
scores_X = cross_val_score(algoX, xtrain, ytrain, cv=5)
print("Mean cross-validation score of XGBR: %.2f" % scores_X.mean())
```

```python
kfold = KFold(n_splits=10, shuffle=True)
kf_cv_scores_R = cross_val_score(algoR, xtrain, ytrain, cv=kfold )
print("K-fold CV average score of LR: %.2f" % kf_cv_scores_R.mean())

kfold = KFold(n_splits=10, shuffle=True)
kf_cv_scores_X = cross_val_score(algoX, xtrain, ytrain, cv=kfold )
print("K-fold CV average score of XGBR: %.2f" % kf_cv_scores_X.mean())
```

```python
ypred_R = algoR.predict(xtest)
mse = mean_squared_error(ytest, ypred_R)
print("MSE of Linear Regression: %.2f" % mse)
print("RMSE of Linear Regression: %.2f" % (mse**(1/2.0)))

ypred_X = algoX.predict(xtest)
mse = mean_squared_error(ytest, ypred_X)
print("MSE of XGBREGRESSOR: %.2f" % mse)
print("RMSE of XGBREGRESSOR: %.2f" % (mse**(1/2.0)))

```
